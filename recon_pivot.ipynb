```python
import pandas as pd
```

```python
# Load the original CSV
df = pd.read_csv('/Users/wagely/Downloads/ID Payout level Jan latest.csv')
```

```python
df.head()
```

```python
pivot_table = pd.pivot_table(df, 
                             values=['outstanding_principal','outstanding_non_principal','outstanding_non_principal_with_tax'],  # Multiple columns
                             index='days_past_due_category',            # Row categories
                             # columns='reimbursement_notice_id',           # Column categories
                             aggfunc='sum')               # Aggregation function


```

```python
# Display the pivot table
print(pivot_table)
```

```python
# Save to CSV
pivot_table.to_csv('pivot_output.csv')
```
